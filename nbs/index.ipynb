# ML Deploy Reference — System Overview


This is the **home page** for the notebook-first ML deployment reference.

It explains the overall structure we are building: infrastructure, notebook-first workflow, and end-to-end traceability.


## What we are building

A Python-first, Linux-only reference platform for ML engineering with:

- **Notebook-first development** via nbdev
- **MLflow traceability** backed by PostgreSQL + S3
- **Distributed compute on Lambda.ai Slurm**
- **Platform services on AWS Kubernetes**
- **Infrastructure via python-terraform**


## Overall architecture

```{mermaid}
flowchart TB
    ENG[ML Engineer] --> NB[Notebook revision in Git\n(immutable source)]
    NB --> UI[Notebook Web UI\ntrigger + visibility]
    UI --> ORCH[Execution orchestration\npolicy + contract checks]

    ORCH --> LOCAL[Local parity execution]
    ORCH --> SLRM[Lambda.ai Slurm execution]
    ORCH --> K8S[AWS Kubernetes execution]

    LOCAL --> MLFLOW[MLflow tracking + registry]
    SLRM --> MLFLOW
    K8S --> MLFLOW

    MLFLOW --> PG[(PostgreSQL backend)]
    MLFLOW --> S3[(S3 / MinIO artifacts)]

    MLFLOW --> OBS[Observability + cost\n(Evidently, Prometheus, Grafana, AWS/Lambda attribution)]
    OBS --> ENG
```


## Notebook-first structure

### Source of truth

- Notebooks in `nbs/` define both narrative and runtime logic.
- Exported modules in `ml_deploy/` are generated artifacts.

### Core notebooks

- [Infrastructure overview](infrastructure_overview.html)
- [Web UI contracts](webui_contracts.html)
- [Vertical slice](vertical_slice.html)
- [MLflow parity](mlflow_parity.html)
- [Execution backends](execution_backends.html)
- [Notebook intake](notebook_intake.html)
- [Terraform bootstrap](terraform_bootstrap.html)


## Traceability spine

Every run should be reconstructable across:

1. **Notebook revision** (immutable Git ref)
2. **Execution backend context** (local / Slurm / Kubernetes)
3. **MLflow run + artifacts** (PostgreSQL + S3 lineage)
4. **Model/deployment records**
5. **Operational metrics + cost attribution**


## Start here next

- For architecture briefing: **Infrastructure overview**
- For interaction contract: **Web UI contracts**
- For executable baseline: **Vertical slice**
- For platform wiring: **Execution backends + Terraform bootstrap**
